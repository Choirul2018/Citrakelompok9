```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Baca gambar 
img_bgr = cv2.imread('3.jpg')  
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

height, width, _ = img_rgb.shape
print(f"Ukuran gambar: {width} x {height}")
```

```python
# gambar sebelum dikonversi
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.title("Gambar Asli")
plt.axis('off')
plt.show()
```

```python
# Definisi Pallet warna 
palette = [
    [0, 0, 0],         # 0 - Hitam
    [255, 0, 0],       # 1 - Merah
    [0, 255, 0],       # 2 - Hijau
    [0, 0, 255],       # 3 - Biru 
]
```

```python
#fungsi untuk menghitung jarak Euclidean antara dua warna RGB
def Euclidean_distance(c1, c2):
    return (c1[0]-c2[0])**2 + (c1[1]-c2[1])**2 + (c1[2]-c2[2])**2
```

```python
# list menyimpan indeks warna 
index_img = []
```

```python
# piksel dalam gambar 
for y in range(height):
    row = []
    for x in range(width):
        r, g, b = img_rgb[y][x]

        min_dist = float('inf')
        closest_index = 0

# membandingkan warna dalam palet
        for i, color in enumerate(palette):
            dist = Euclidean_distance([r, g, b], color)
            if dist < min_dist:
                min_dist = dist
                closest_index = i
        row.append(closest_index)
        
    index_img.append(row)

index_img = np.array(index_img)
```

```python
# hasil konversi terindeks 
plt.figure(figsize=(10, 10))
plt.imshow(index_img, cmap='tab10')
plt.title("Citra Terindeks (Euclidean Distance)")
plt.show()
```
