```python
from PIL import Image
import math
import matplotlib.pyplot as plt
```

```python
def convert_to_grayscale(pixel):
    r, g, b = pixel

    gray = int(0.299 * r + 0.587 * g + 0.114 * b)

    return gray
```

```python
def convert_to_binary(gray, threshold=230):
    if gray >= threshold :
        binary = 1
    else :
        binary = 0
    return binary
```

```python
# Buka gambar
img = Image.open('UNP.png')
img = img.convert('RGB')
width, height = img.size
```

```python
plt.imshow(img)
plt.axis('off')
plt.title('Gambar Asli')
plt.show()
```

```python
result_matrix = []
```

```python
for y in range(height):
    row = []
    for x in range(width):
        pixel = img.getpixel((x, y))
        gray = convert_to_grayscale(pixel)
        binary_value = convert_to_binary(gray, threshold=100)
        row.append(binary_value)
    result_matrix.append(row)
```

```python
# Buat gambar baru dari hasil
result_img = Image.new('1', (width, height))

for y in range(height):
    for x in range(width):
        color = result_matrix[y][x]
        result_img.putpixel((x, y), color)
```

```python
# Tampilkan gambar hasil menggunakan matplotlib
plt.imshow(result_img)
plt.axis('off')
plt.title('Hasil Konversi Warna')
plt.show()
```
